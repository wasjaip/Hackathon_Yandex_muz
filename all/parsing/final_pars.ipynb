```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Указание URL
url = 'https://www.deezer.com/track/453120892'

# Запрос к странице
response = requests.get(url)

# Проверка на успешный ответ (код 200)
if response.status_code == 200:
    # Парсинг HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Поиск тега с названием трека
    title_tag = soup.find('title')

    if title_tag:
        # Извлечение текста из тега
        title_text = title_tag.text

        # Вывод названия трека
        print(title_text)
else:
    print('Failed to retrieve the page: ', response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

# Указание URL
url = 'https://www.deezer.com/track/116348128'

# Запрос к странице
response = requests.get(url)

# Проверка на успешный ответ (код 200)
if response.status_code == 200:
    # Парсинг HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Поиск тега с названием трека
    title_tag = soup.find('title')

    if title_tag:
        # Извлечение текста из тега
        title_text = title_tag.text

        # Вывод названия трека
        print('Title:', title_text)
    else:
        print('Не удалось найти тег title.')

    # Поиск первого тега div
    first_div = soup.find('div', text='Şarkı sözleri')

    if first_div:
        # Поиск следующего соседнего тега div
        lyrics_div = first_div.find_next_sibling('div')

        if lyrics_div:
            # Извлечение текста из тега
            lyrics_text = lyrics_div.text

            # Вывод текста песни
            print('Lyrics:', lyrics_text)
        else:
            print('Не удалось найти div с текстом песни.')
    else:
        print('Не удалось найти первый div.')
else:
    print('Не удалось получить страницу: ', response.status_code)
```

### Цикл от 1 до 101 включительно ( желательно при большом обьеме рабить на итерации по 20 или 10 файлов)

ниже будут вариации названия, изменения касается только названия полей по разным vpn

```python


for i in range(1, 20):  # Цикл от 1 до 101 включительно
    # Имя файла для чтения
    input_file = f'output{i}.csv'

    # Загрузка данных из CSV файла
    data_df = pd.read_csv(input_file)

    # Список для хранения данных
    result_list = []

    # Перебор каждой строки в DataFrame
    for index, row in data_df.iterrows():
        # Указание URL
        url = row['deezer_url']

        # Проверка на отсутствие данных (nan)
        if pd.isna(url):
            print(f'Missing URL for index {index}, ISRC: {row["isrc"]}')
            continue  # Пропуск текущей итерации и переход к следующей строке DataFrame

        # Запрос к странице
        response = requests.get(url)

        # Проверка на успешный ответ (код 200)
        if response.status_code == 200:
            # Парсинг HTML
            soup = BeautifulSoup(response.content, 'html.parser')

            # Поиск тега с названием трека
            title_tag = soup.find('title')
            first_div = soup.find('div', text='Lyrics')

            if first_div:
                # Поиск следующего соседнего тега div
                lyrics_div = first_div.find_next_sibling('div')

                if lyrics_div:
                    # Извлечение текста из тега
                    lyrics_text = lyrics_div.text.strip()  # Удаление лишних пробелов и переносов строки
                else:
                    lyrics_text = 'текст не найден'
            else:
                lyrics_text = 'текст не найден'

            # Извлечение текста из тега title
            title_text = title_tag.text if title_tag else 'Title not found'

            # Добавление данных в список
            result_list.append({
                'isrc': row['isrc'],
                'title': title_text,
                'lyrics': lyrics_text
            })
        else:
            print(f'Failed to retrieve the page: {response.status_code}, URL: {url}')

    # Создание нового DataFrame из списка словарей
    result_df = pd.DataFrame(result_list)

    # Удаление всего, что находится после знака ":"
    result_df['title'] = result_df['title'].str.split(':').str[0]

    # Добавление поля "cav", которое содержит 1, если в title есть слово "Cover", и 0 в противном случае
    result_df['cav'] = result_df['title'].str.contains('Cover', case=False).astype(int)

    # Имя файла для сохранения результатов
    output_file = f'result_n{i}.csv'

    # Сохранение result_df в CSV файл
    result_df.to_csv(output_file, index=False)
```
